```python
import plotly.graph_objects as go
import numpy as np

# Define thresholds and parameters
stable_threshold = 15  # Max deviation for stability in degrees
jibe_angle_diff = 30  # Approximate angle change for a jibe
tack_angle_diff = 120  # Approximate angle change for a tack
min_samples_stable = 5  # Minimum samples for stability

# Function to identify stable regions in COG
def find_stable_regions(cog_series, threshold, min_samples):
    stable_regions = []
    current_region = [0]

    for i in range(1, len(cog_series)):
        if abs(cog_series[i] - cog_series[current_region[0]]) <= threshold:
            current_region.append(i)
        else:
            if len(current_region) >= min_samples:
                stable_regions.append(current_region)
            current_region = [i]

    if len(current_region) >= min_samples:
        stable_regions.append(current_region)

    return stable_regions

# Function to identify turns based on stable regions
def identify_turns(stable_regions, cog_series):
    turns = []

    for i in range(len(stable_regions) - 1):
        start_region = stable_regions[i]
        end_region = stable_regions[i + 1]

        # Calculate the stable COG values
        stable_start = np.mean(cog_series[start_region])
        stable_end = np.mean(cog_series[end_region])

        # Calculate the angle difference
        angle_diff = abs(stable_end - stable_start)
        if angle_diff > 180:
            angle_diff = 360 - angle_diff

        # Classify the turn
        if jibe_angle_diff - 25 <= angle_diff <= jibe_angle_diff + 25:
            turn_type = "jibe"
        elif tack_angle_diff - 20 <= angle_diff <= tack_angle_diff + 20:
            turn_type = "tack"
        else:
            continue

        # Identify the midpoint of the turn
        midpoint_index = start_region[-1] + (end_region[0] - start_region[-1]) // 2
        midpoint_cog = cog_series[midpoint_index]

        # Add the turn details
        turns.append({
            "start_cog": stable_start,
            "end_cog": stable_end,
            "angle_diff": angle_diff,
            "turn_type": turn_type,
            "midpoint_index": midpoint_index,
            "midpoint_cog": midpoint_cog
        })

    return turns

# Example usage
cog_series = sailmon_cut['COG - Course over Ground'].to_numpy()

# Step 1: Find stable regions
stable_regions = find_stable_regions(cog_series, stable_threshold, min_samples_stable)

# Step 2: Identify turns
turns = identify_turns(stable_regions, cog_series)

# Plot the results
fig = go.Figure()

# Add the full trajectory
fig.add_trace(go.Scatter(
    x=sailmon_cut['longitude'],
    y=sailmon_cut['latitude'],
    mode='lines',
    line=dict(color='lightgray'),
    name='Full Trajectory'
))

# Add turn points
for turn in turns:
    turn_color = "red" if turn["turn_type"] == "jibe" else "blue"
    fig.add_trace(go.Scatter(
        x=[sailmon_cut.iloc[turn['midpoint_index']]['longitude']],
        y=[sailmon_cut.iloc[turn['midpoint_index']]['latitude']],
        mode='markers',
        marker=dict(color=turn_color, size=12),
        name=turn["turn_type"].capitalize(),
        hovertext=(
            f"Turn Type: {turn['turn_type']}<br>"
            f"Start COG: {turn['start_cog']:.2f}°<br>"
            f"End COG: {turn['end_cog']:.2f}°<br>"
            f"Angle Difference: {turn['angle_diff']:.2f}°<br>"
            f"Midpoint COG: {turn['midpoint_cog']:.2f}°"
        ),
        hoverinfo="text"
    ))

# Customize the layout
fig.update_layout(
    title="Detected Turns with Details",
    xaxis_title="Longitude",
    yaxis_title="Latitude",
    legend_title="Turn Type",
    template="plotly_white"
)

# Add stable (straight) segments to the plot
for region in stable_regions:
    region_indices = region
    fig.add_trace(go.Scatter(
        x=sailmon_cut.iloc[region_indices]['longitude'],
        y=sailmon_cut.iloc[region_indices]['latitude'],
        mode='lines',
        line=dict(color='green', width=3),
        name='Stable Segment',
        showlegend=False,  # Avoid legend repetition
        hoverinfo='skip'  # Optional: don't show hover for every point
    ))

# Show the plot
fig.show()
```

```python
# ---------- פונקציות עזר ----------

# חישוב זווית בין שני שיפועים
def angle_between_lines(m1, m2):
    angle_rad = np.arctan(np.abs((m1 - m2) / (1 + m1 * m2)))
    return np.degrees(angle_rad)

# התאמת קו ישר עם RANSAC
def fit_line_ransac(x, y):
    model = RANSACRegressor(random_state=0)
    x = x.reshape(-1, 1)
    model.fit(x, y)
    return model.estimator_.coef_[0]  # השיפוע של הקו
```

```python
# ---------- הגדרות ----------

window = 10  # מספר נקודות לפני ואחרי הפנייה
min_angle = 25
max_angle = 110
df_turns['df_index'] = df_turns.index

# df: טבלת נתיב ראשית (מכילה latitude, longitude)
# turns_df: טבלה עם peak_time של הפניות
# נניח שכבר הפכנו את עמודות הזמנים לתאריכים:
```

```python
# ---------- חישוב ואימות פניות ----------

confirmed_turns = []
rejected_turns = []

for _, row in df_turns.iterrows():
    turn_index = row['turn_index']
    idx = row['df_index']
    
    if idx - window < 0 or idx + window >= len(traj_df):
        rejected_turns.append((idx, "Out of bounds"))
        continue

    # קטעים לפני ואחרי הפנייה
    seg1 = traj_df.iloc[idx - window:idx]
    seg2 = traj_df.iloc[idx + 1:idx + 1 + window]

    # שימוש ב-longitude כ-x ו-latitude כ-y
    x1, y1 = seg1['longitude'].values, seg1['latitude'].values
    x2, y2 = seg2['longitude'].values, seg2['latitude'].values

    try:
        m1 = fit_line_ransac(x1, y1)
        m2 = fit_line_ransac(x2, y2)
        angle = angle_between_lines(m1, m2)
    except Exception as e:
        rejected_turns.append((turn_index, idx, f"Fitting error: {str(e)}"))
        continue

    if angle < min_angle:
        rejected_turns.append((turn_index, idx, f"Angle too small: {angle:.2f}°"))
    elif angle > max_angle:
        rejected_turns.append((turn_index, idx, f"Angle too large: {angle:.2f}°"))
    else:
        confirmed_turns.append((turn_index, idx, angle))
```

```python
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter('ignore', np.RankWarning)

# הוספת קואורדינטות לנקודות החשודות
inlier_turns['longitude'] = inlier_turns['df_index'].apply(lambda i: sailmon_cut.loc[i, 'longitude'])
inlier_turns['latitude'] = inlier_turns['df_index'].apply(lambda i: sailmon_cut.loc[i, 'latitude'])

# הוספת קואורדינטות לכל קבוצה
groups_with_coords = []
for group in groups:
    group = group.copy()
    group['longitude'] = group['df_index'].apply(lambda i: sailmon_cut.loc[i, 'longitude'])
    group['latitude'] = group['df_index'].apply(lambda i: sailmon_cut.loc[i, 'latitude'])
    groups_with_coords.append(group)

# מציאת הנקודה החשודה הקרובה ביותר לאפקס של פרבולה שמותאמת לנתיב המלא
selected_apex_points = []
for group in groups_with_coords:
    if len(group) == 0:
        continue

    idx_min = group['df_index'].min()
    idx_max = group['df_index'].max()
    
    # הרחבת הטווח ל־2 דגימות לפני ו־2 אחרי
    start_idx = max(0, idx_min - 2)
    end_idx = min(len(sailmon_cut) - 1, idx_max + 2)
    segment = sailmon_cut.iloc[start_idx:end_idx + 1]

    x = segment['longitude'].values
    y = segment['latitude'].values

    if len(x) < 3:
        continue

    try:
        coeffs = np.polyfit(x, y, 2)
        x_apex = -coeffs[1] / (2 * coeffs[0])
        y_apex = np.polyval(coeffs, x_apex)

        # חישוב מרחק מה-apex לכל נקודה חשודה בקבוצה
        distances = np.sqrt((group['longitude'] - x_apex)**2 + (group['latitude'] - y_apex)**2)
        closest_idx = distances.idxmin()
        selected_apex_points.append(group.loc[closest_idx])
    except Exception as e:
        continue  # אם ההתאמה נכשלה, מדלגים

# יצירת טבלה עם הנקודות שנבחרו
selected_apex_df = pd.DataFrame(selected_apex_points).reset_index(drop=True)


# נקודות אפקס שנבחרו
fig.add_trace(go.Scatter(
    x=selected_apex_df['longitude'],
    y=selected_apex_df['latitude'],
    mode='markers',
    marker=dict(color='red', size=7, symbol='x'),
    name='Selected Apex Points',
    hovertext=[
        f"Turn Group: {row['group_id']}<br>"
        f"Turn Index: {row['turn_index']}<br>"
        f"DF Index: {row['df_index']}<br>"
        f"Angle: {row['Angle (degrees)']:.2f}°<br>"
        f"COG_diff: {row['COG_diff']:.2f}°"
        for _, row in selected_apex_df.iterrows()
    ],
    hoverinfo='text'
))


fig.update_layout(
    title="Selected Turn Points Closest to Apex from Parabolic Fit",
    xaxis_title="Longitude",
    yaxis_title="Latitude",
    template="plotly_white"
)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import RANSACRegressor
from sklearn.metrics import mean_squared_error

def fit_line_and_error(x, y):
    """מתאים קו בעזרת RANSAC ומחזיר גם שגיאה (MSE) וגם השיפוע"""
    model = RANSACRegressor(random_state=0)
    model.fit(x.reshape(-1, 1), y)
    y_pred = model.predict(x.reshape(-1, 1))
    error = mean_squared_error(y, y_pred)
    slope = model.estimator_.coef_[0]
    return error, slope

def find_straight_segment(traj_df, center_idx, direction='backward', max_window=30, min_samples=5, error_threshold=1e-6):
    """
    הולך אחורה/קדימה מהשיא עד שהשגיאה של ההתאמה לקו יורדת מתחת לסף
    מחזיר את האינדקס שבו מתחיל (או מסתיים) אזור ישר
    """
    if direction == 'backward':
        indices = range(center_idx - min_samples, max(center_idx - max_window, 0), -1)
    else:  # forward
        indices = range(center_idx + min_samples, min(center_idx + max_window, len(traj_df)))

    for i in indices:
        if direction == 'backward':
            x = traj_df['longitude'].iloc[i:center_idx].values
            y = traj_df['latitude'].iloc[i:center_idx].values
            start_idx = i
        else:
            x = traj_df['longitude'].iloc[center_idx:i].values
            y = traj_df['latitude'].iloc[center_idx:i].values
            start_idx = i

        if len(x) < min_samples:
            continue

        error, slope = fit_line_and_error(np.array(x), np.array(y))
        if error < error_threshold:
            return start_idx

    return None

def apply_turn_detection_by_good_fit(inlier_turns, traj_df, max_window=30, error_threshold=1e-6):
    """
    מפעיל על כל שיא פנייה:
    - מוצא התחלה (אחורה) וסיום (קדימה) לפי שיפור בהתאמה לקו ישר
    """
    results = []

    for _, row in inlier_turns.iterrows():
        center_idx = row['df_index']

        if center_idx < max_window or center_idx >= len(traj_df) - max_window:
            continue  # לא מספיק נתונים סביב הפנייה

        start_idx = find_straight_segment(traj_df, center_idx, direction='backward', max_window=max_window, error_threshold=error_threshold)
        end_idx = find_straight_segment(traj_df, center_idx, direction='forward', max_window=max_window, error_threshold=error_threshold)

        if start_idx is not None and end_idx is not None:
            results.append({
                'turn_index': row['turn_index'],
                'df_index': center_idx,
                'start_idx': start_idx,
                'end_idx': end_idx
            })

    return pd.DataFrame(results)
```

```python
def create_turns_summary_table(turn_boundaries_df, inlier_turns, traj_df):
    """
    מחזיר DataFrame בפורמט:
    "",turn_index,turn_type,start_time,peak_time,end_time
    כשהאינדקס הוא df_index (peak_index)
    והזמנים הם datetime מלא
    """

    # המרה ל-datetime אם עדיין לא בוצעה
    traj_df['timestamp'] = pd.to_datetime(traj_df['time'])

    # מיזוג עם סוג הפנייה
    merged = turn_boundaries_df.merge(
        inlier_turns[['turn_index', 'turn_type']], on='turn_index', how='left'
    )

    # שליפת זמני התחלה, שיא, וסיום לפי האינדקסים
    merged['start_time'] = merged['start_idx'].apply(lambda idx: traj_df.loc[idx, 'timestamp'])
    merged['peak_time'] = merged['df_index'].apply(lambda idx: traj_df.loc[idx, 'timestamp'])
    merged['end_time'] = merged['end_idx'].apply(lambda idx: traj_df.loc[idx, 'timestamp'])
    
    # הגדרת df_index כאינדקס
    merged.set_index('df_index', inplace=True)

    # בחירת עמודות מסודרת
    summary_df = merged[['turn_index', 'turn_type', 'start_time', 'peak_time', 'end_time','turn_index_update']].copy()

    # שם עמודת אינדקס - "" כמו בדוגמה שלך
    summary_df.index.name = '""'

    return summary_df
```
